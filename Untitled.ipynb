```python
from PIL import Image
import numpy as np

def preprocess_image(image_path):
    img = Image.open(image_path)
    img = img.resize((224, 224))  # Resize image to 224x224 pixels
    img = np.array(img) / 255.0  # Normalize pixel values
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model

# Load ResNet50 model + higher level layers
base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

def extract_features(img_path, model):
    img = image.load_img(img_path, target_size=(224, 224))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    features = model.predict(img_data)
    return features.flatten()

print("done")
```

```python
import os
image_dir = r"Eid_edit"
    
# List all image files in the directory
image_files = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.jpg')]
print(image_files)
```

```python
from sklearn.cluster import KMeans

def cluster_images(features_array, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(features_array)
    return kmeans.labels_

# Example usage
image_paths = image_files
features_list = [extract_features(img_path, base_model) for img_path in image_paths]
features_array = np.array(features_list)
labels = cluster_images(features_array)
```

```python
from sklearn.neighbors import NearestNeighbors

def find_similar_images(user_img_path, image_paths, model, n_neighbors=3):
    # Extract features from all images in the dataset
    features_list = [extract_features(img_path, model) for img_path in image_paths]
    features_array = np.array(features_list)
    
    # Fit NearestNeighbors model
    nbrs = NearestNeighbors(n_neighbors=n_neighbors, algorithm='ball_tree').fit(features_array)
    
    # Extract features from the user input image
    user_features = extract_features(user_img_path, model).reshape(1, -1)
    
    # Find similar images
    distances, indices = nbrs.kneighbors(user_features)
    
    # Retrieve similar images
    similar_images = [image_paths[i] for i in indices[0]]
    return similar_images

# Example usage
user_img_path = 'Eid_edit/image_29.jpg'
similar_images = find_similar_images(user_img_path, image_paths, base_model)
print("Similar images:", similar_images)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

def display_images_in_grid(image_paths, grid_size=(2, 3)):
    """
    Display images in a grid.
    
    Parameters:
    - image_paths: list of image file paths
    - grid_size: tuple, the grid size (rows, columns)
    """
    rows, cols = grid_size
    fig, axes = plt.subplots(rows, cols, figsize=(15, 10))

    for i, img_path in enumerate(image_paths):
        row = i // cols
        col = i % cols
        if row < rows:
            img = Image.open(img_path)
            axes[row, col].imshow(img)
            axes[row, col].axis('off')  # Hide the axes

    # Hide any remaining axes if there are fewer images than grid cells
    for j in range(i + 1, rows * cols):
        row = j // cols
        col = j % cols
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()

# Example usage
# image_paths = ['Eid_edit/image_2.jpg', 'Eid_edit\\image_2.jpg', 'Eid_edit\\image_43.jpg', 'Eid_edit\\image_34.jpg', 'Eid_edit\\image_38.jpg', 'Eid_edit\\image_33.jpg']
display_images_in_grid(similar_images, grid_size=(2, 3))
```
